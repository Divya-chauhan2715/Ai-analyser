## 6️⃣ Recommendation System

We build two types of recommendation engines:
1. **Content-Based Filtering** — Recommends products based on what categories a customer likes
2. **Collaborative Filtering** — Recommends products based on what similar customers bought
3. **Hybrid** — Combines both approaches for better results

```python
# Run the recommendation system demo
hybrid_recommender = demo_recommendations(df, n_customers=3, n_recs=5)
```

```python
# Try recommending for a specific customer
# Get a customer with many orders for demo
customer_orders = df.groupby('Customer_ID')['Order_ID'].nunique().sort_values(ascending=False)
test_customer = customer_orders.index[0]

print(f'\n🔍 Testing recommendations for: {test_customer}')
print(f'   (This customer has {customer_orders.iloc[0]} orders)\n')

# Content-based
content_rec = ContentBasedRecommender(df)
print('📋 Content-Based Recommendations:')
content_rec.recommend(test_customer, n=5)
```

```python
# Collaborative
collab_rec = CollaborativeRecommender(df)

print('\n🤝 Similar Users:')
display(collab_rec.get_similar_users(test_customer, n=5))

print('\n🤝 Collaborative Filtering Recommendations:')
collab_rec.recommend(test_customer, n=5)
```

```python
# Hybrid recommendations
hybrid = HybridRecommender(df)
print('⭐ Hybrid Recommendations (Best of Both):')
hybrid.recommend(test_customer, n=5)
```

### How the Recommendation System Works:

**Content-Based Filtering:**
- Looks at what categories/sub-categories a customer has bought from
- Scores products based on category similarity to their purchase history
- Higher score = more relevant to customer's preferences

**Collaborative Filtering:**
- Builds a User-Item matrix (who bought what)
- Uses Cosine Similarity to find similar users
- Recommends products that similar users bought but this customer hasn't

**Hybrid:**
- Combines both scores (40% content + 60% collaborative)
- Provides the most robust recommendations
